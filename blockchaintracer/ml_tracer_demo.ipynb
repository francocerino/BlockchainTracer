# BlockchainTracer Demo

```python
from blockchain_tracer import BlockchainTracer
```

```python
provider_url = 'https://mainnet.infura.io/v3/3419054254da4a54963cf2e6eea61199'
bc_tracer = BlockchainTracer(provider_url=provider_url)
```

```python
bc_tracer.get_data()
```

```python
bc_tracer.update_data(key = 123)
```

```python
bc_tracer.get_data()
```

```python
data_hash = bc_tracer.compute_hash('message')
```

```python
bc_tracer.update_data(data_hash = data_hash)
```

```python
bc_tracer._blockchain_data
```

```python
bc_tracer.compute_hash(bc_tracer._blockchain_data)
```

```python
bc_tracer.write_to_blockchain(only_write_hash=False)
```

ahora vemos los datos escritos en la blockchain

```python
dict_info_tx, tx_input = bc_tracer.get_transaction_details(tx_hash)
dict_info_tx, tx_input
```

```python
data_saved = dict_info_tx['transaction']['data']
data_saved
```

# MLTracer Demo

```python
from ml_tracer import MLTracer
```

```python
# Instantiate MLTracer (provider_url is optional for local operations)
ml_tracer = MLTracer()
```

```python
ml_tracer.get_model_card
```

```python
# Show available model card fields
print("Model card fields:", ml_tracer.model_card_fields)
```

```python
# Update the model card incrementally
kwargs = {'model_name':"MyModel",
          'library_name':"transformers"
         }
ml_tracer.update_model_card(**kwargs)
```
